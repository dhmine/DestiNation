```python
def open_index_file(feed_name):

    """
    open the index file in given feed
    :param feed_name:
    :return: json object
    """

    path = F'./data/{feed_name}/index.json'

    with open(path) as f:
        feed_data = json.load(f)

    return feed_data
```

```python
def get_label_file(item):
    """
    get the even label and from an item
    :param item: item form index json file
    :return: tupile (label, file)
    """
    return item['label'], item['file']
```

```python
import folium
def create_map(df):


    m = folium.Map(location=[df.latitude.max(), df.latitude.max()], zoom_start=4, tiles='OpenStreetMap')
    for _, row in df.iterrows():

        folium.CircleMarker(
            location= [row['latitude'], row['longitude']],
            radius =8,

            popup = F' label:  {row["label"]} \n  '+
                    F'comment: {row["comment"]}  ',

        ).add_to(m)
    return m
```
